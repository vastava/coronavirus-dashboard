```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df = pd.read_csv('corona_country_daily_v3.csv')
df
```

```python
countries = np.array(df.columns[1:])
```

```python
def spread_and_fit(diffs, date, tempdict):
    for x in diffs:
        date_parsed = datetime.strptime(date, "%m/%d/%y")
        str_date = (date_parsed - timedelta(days=int(x))).strftime("%-m/%-d/%y")
        if str_date in tempdict:        
            tempdict[str_date] += 1
        else:
            tempdict[str_date] = 1
    return tempdict
```

```python
def calc_diffs(size):
    np.random.seed(42)
    diffs = np.random.normal(10, 3, int(abs(size)))
    return np.array([int(x) for x in diffs])
```

```python
mega_dict = {}
for country in countries:
    mega_dict[country] = {}
    for index, row in df.iterrows():
        diffs = calc_diffs(row[country])        
        mega_dict[country] = spread_and_fit(diffs, row["date"], mega_dict[country])
```

```python
mega_dict
```

```python
onset = pd.DataFrame.from_dict(mega_dict, orient='index').fillna(0).transpose()
onset
```

```python
onset.columns = [str(col) + '_onset' for col in onset.columns]
onset
```

```python
onset = onset.reset_index().rename(columns={'index':'date'})
onset
```

```python
df.merge(onset, left_on="date", right_on="date", how='outer').fillna(0).to_csv('world_onset_v3.csv', index=False)
```
